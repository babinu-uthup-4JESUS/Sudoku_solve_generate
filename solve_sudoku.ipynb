```python
def get_sub_square_row_col(i):
    return 3*((int)(i/3)) + 1
```

```python
def get_sub_square_elem_indices(i, j, input_arr=GIVEN_INPUT):
    row = get_sub_square_row_col(i)
    col = get_sub_square_row_col(j)
    cur_list = list()
    for i1 in range(row - 1, row + 2):
        for j1 in range(col - 1, col + 2):
            cur_list.append((i1, j1))
    
    return cur_list
```

```python
def get_sub_square_all_elem(i, j, input_arr=GIVEN_INPUT):
    row = get_sub_square_row_col(i)
    col = get_sub_square_row_col(j)
    cur_list = list()
    indices_list = get_sub_square_elem_indices(i, j, input_arr)
    for index in indices_list:
        i1 = index[0]
        j1 = index[1]
        cur_list.append(input_arr[i1, j1])        

    return cur_list
```

```python
def get_sub_square_elem(i, j, input_arr=GIVEN_INPUT):
    cur_list = get_sub_square_all_elem(i, j, input_arr)
    return set(cur_list)
```

```python
def get_row_indices(i, input_arr=GIVEN_INPUT):
    return list([(i, j) for j in range(9)])
```

```python
def get_row_all_elem(i, input_arr=GIVEN_INPUT):
    indices_list = get_row_indices(i, input_arr)
    cur_list = list()
    for index in indices_list:
        i1 = index[0]
        j1 = index[1]
        cur_list.append(input_arr[i1, j1])        

    return cur_list
```

```python
def get_row_elem(i, input_arr=GIVEN_INPUT):
    cur_list = get_row_all_elem(i, input_arr)
    return set(cur_list)    
```

```python
def get_col_indices(j, input_arr=GIVEN_INPUT):
    return list([(i, j) for i in range(9)])
```

```python
def get_col_all_elem(j, input_arr=GIVEN_INPUT):
    indices_list = get_col_indices(j, input_arr)
    cur_list = list()
    for index in indices_list:
        i1 = index[0]
        j1 = index[1]
        cur_list.append(input_arr[i1, j1])        

    return cur_list
```

```python
def get_col_elem(j, input_arr=GIVEN_INPUT):
    cur_list = get_col_all_elem(j, input_arr)
    return set(cur_list)    
```

```python
def get_all_rel_indices_elem(i, j, input_arr=GIVEN_INPUT):
    row_indices = get_row_indices(i, input_arr)
    col_indices = get_col_indices(j, input_arr)
    sub_square_indices = get_sub_square_elem_indices(i, j, input_arr)
    all_rel_indices = set(row_indices + col_indices + sub_square_indices)
    return all_rel_indices
```

```python
def get_row_col_sub_square_elem(i, j, input_arr=GIVEN_INPUT):
    row_elem = get_row_elem(i, input_arr)
    col_elem = get_col_elem(j, input_arr)
    sub_square_elem = get_sub_square_elem(i, j, input_arr)
    
    union_elem = row_elem | col_elem | sub_square_elem
    
    return union_elem
```

```python
# We count the number of non zero elements in the same row, column, or sub square of the current element.
# This  value is used to fine tune our algorithm later.
def get_num_known_elem_vicinity_cur_elem(i, j, input_arr=GIVEN_INPUT):
    rel_indices = get_all_rel_indices_elem(i, j, input_arr)
    count = 0
    for index in rel_indices:
        (i1, j1) = index
        if input_arr[i1, j1] != 0:
            count += 1
    return count
```

```python
class explorationState():
    """
        The class which stores the current state of the array as well as the element which is to be explored
    """
    
    
    def __init__(self, row_index, col_index, given_elem, elem_impact, cur_array):
        self.__row_index = row_index
        self.__col_index = col_index
        self.__given_elem = given_elem
        self.__elem_impact = elem_impact
        self.__cur_array = cur_array        

    def get_elements(self):
        return (self.__row_index, self.__col_index, self.__given_elem, self.__elem_impact, self.__cur_array)
    
    def display(self):
        print(self.__row_index, self.__col_index, self.__given_elem, self.__elem_impact, self.__cur_array)    
```

```python
def check_repetitive_non_zero_elem(input_array):
    repetitive_non_zero_elem = False
    non_zero_elem = [x for x in input_array if x != 0]
    if len(non_zero_elem) != len(set(non_zero_elem)):
        repetitive_non_zero_elem = True
     
    return repetitive_non_zero_elem
```

```python
def check_if_state_is_correct(output_arr):
    state_correct = True
    for i in range(9):
        row = int(i/3)
        col = i%3

        if (check_repetitive_non_zero_elem(get_row_all_elem(i, input_arr=output_arr)) or 
            check_repetitive_non_zero_elem(get_col_all_elem(i, input_arr=output_arr)) or
            check_repetitive_non_zero_elem(get_sub_square_all_elem(3*row, 3*col, input_arr=output_arr))):
            state_correct = False
            break

    return state_correct
        
```

```python
# The function takes as input the following:
# (a) The array showing the number of known elements in the vicinity for each position --- array_1
# (b) The array showing the set of possible values for each for position --- array_2
# and returns the following:
# (a) The position (that is the row and column) of the maximum element in array_1, whose corresponding entry in 
#     array_2 has a length of 0 or greater than 1 --- output_1
# (b) The corresponding element in array_2 for the position obtained in output_1
# (c) The corresponding element in array_1 for the position obtained in output_1
def get_max_num_known_elem_row_col(num_known_elem_arr, possible_elems):
    sorted_array = -np.sort(-num_known_elem_arr, axis=None)
    output_row = -1
    output_col = -1
    output_set_vals = set()
    required_elem_found = False
    for elem in sorted_array:
        indices = np.where(num_known_elem_arr == elem)
        row_indices = indices[0]
        col_indices = indices[1]
        assert(len(row_indices) == len(col_indices))
        for i in range(len(row_indices)):
            row = row_indices[i]
            col = col_indices[i]
            given_set = possible_elems[row][col]
 
            # IMPORTANT: We have 2 cases to consider here. When length of the set is 0, it means
            # that set cannot take any element, implying that we are in an erroneous state, on which
            # we exit the loop (though we take note of the corresponding state, since the set is empty,
            # it will not be pushed into the stack, implicitly stopping our traversal through that corresponding
            # path) . If the length of the set is 1, it means that we already know the element
            # in this position and it does not make sense to push it to the stack. 
            if ((len(given_set) == 0) or (len(given_set) > 1)):
                output_row = row
                output_col = col
                output_set_vals = possible_elems[row][col]
                required_elem_found = True
                break
        if required_elem_found:
            break
        
    return (output_row, output_col, output_set_vals, elem) 
    
```

```python
# The function takes as input the following:
# (a) The array showing the number of known elements in the vicinity for each position --- array_1
# (b) The array showing the set of possible values for each for position --- array_2
# and returns the following:
# (a) The position (that is the row and column) of the first element in array_1, whose corresponding entry in 
#     array_2 has the least length apart from 1 --- output_1
# (b) The corresponding element in array_2 for the position obtained in output_1
# (c) The corresponding element in array_1 for the position obtained in output_1
def get_min_num_possiblities_first_unknown_elem_row_col(num_known_elem_arr, possible_elems):
    output_row = -1
    output_col = -1
    output_set_vals = set()
    required_elem_found = False

    set_lengths_array  = np.zeros((9,9))
    for i in range(9):
        for j in range(9):
            set_lengths_array[i,j] = len(possible_elems[i][j])
    
    max_set_len = int(np.max(set_lengths_array))
    for length in range(max_set_len + 1):

        # IMPORTANT: We omit the entries having length 1, as these correspond to known elements,
        # which we are not interested in.
        if length == 1:
            continue

        indices = np.where(set_lengths_array == length)
        row_indices = indices[0]
        col_indices = indices[1]
        assert(len(row_indices) == len(col_indices))
        if len(row_indices) > 0:
            output_row = row_indices[0]
            output_col = col_indices[0] 
            num_known_elem = num_known_elem_arr[output_row][output_col]
            output_set_vals = possible_elems[output_row][output_col]
            required_elem_found = True
            break
                
    return (output_row, output_col, output_set_vals, num_known_elem) 
```

```python
# The function takes as input the following:
# (a) The array showing the current state of the array, whose values are being discovered --- array_1
# (b) The array showing the number of known elements in the vicinity for each position --- array_2
# (c) The array showing the set of possible values for each for position --- array_3
# (d) The mode of the program which is to be used.
# and returns a list of which each entry has the following:
# Row, column, corresponding value of the element as well as it's impact depending on the mode used.
def get_unknown_elem_row_col_as_list(output_arr, impact_arr, possible_elems, run_mode='MIN_NUM_POSSIBILITIES_MIN_NUM_KNOWN_ELEM'):
    rel_list = list()
    
    if run_mode == 'FIRST_UNKNOWN':
        (row, col, corr_possible_elems, cur_impact) = get_zero_elem_row_col_num_known_elem(output_arr, possible_elems)
    elif run_mode == 'MIN_NUM_POSSIBILITIES_FIRST_UNKNOWN':
        (row, col, corr_possible_elems, cur_impact) = \
            get_min_num_possiblities_first_unknown_elem_row_col(impact_arr, possible_elems)
    elif run_mode == 'MAX_NUM_KNOWN_ELEM':
        (row, col, corr_possible_elems, cur_impact) = \
            get_max_num_known_elem_row_col(impact_arr, possible_elems)
    elif run_mode == 'MIN_NUM_POSSIBILITIES_MAX_NUM_KNOWN_ELEM':
        (row, col, corr_possible_elems, cur_impact) = \
            get_min_num_possibilities_max_num_known_elem_row_col(impact_arr, possible_elems)
    elif run_mode == 'MIN_NUM_POSSIBILITIES_MIN_NUM_KNOWN_ELEM':
        (row, col, corr_possible_elems, cur_impact) = \
            get_min_num_possibilities_min_num_known_elem_row_col(impact_arr, possible_elems)
    elif run_mode == 'MIN_NUM_KNOWN_ELEM':
        (row, col, corr_possible_elems, cur_impact) = \
            get_min_num_known_elem_row_col(impact_arr, possible_elems)
    else:
        print('Incorrect mode specified. Using mode MIN_NUM_POSSIBILITIES_MIN_NUM_KNOWN_ELEM by default')
        (row, col, corr_possible_elems, cur_impact) = \
            get_min_num_possibilities_min_num_known_elem_row_col(impact_arr, possible_elems)
        
    for elem in corr_possible_elems:
        rel_list.append((row, col, elem, cur_impact))
        
    return rel_list
    
```

```python
def sort_items_by_impact_and_push_to_stack(
    output_arr, impact_arr, possible_elems, stack, run_mode='MIN_NUM_POSSIBILITIES_MIN_NUM_KNOWN_ELEM'):

    rel_items = get_unknown_elem_row_col_as_list(output_arr, impact_arr, possible_elems, run_mode)
        
    for item in rel_items:
        (row, col, elem, impact) = item
        stack.put(explorationState(row, col, elem, impact, output_arr))

```

### Come to the core algorithm

Let us come to the part where the real meat is!


```python
# As the name suggests, we do the obvious analysis for each position of the array, 
# by evaluating all the possiblities of the corresponding row, column and sub square and 
# updating the corresponding position, if only one value was found, which satisifed the constraints
# of Sudoku.
def do_obvious_analysis(input_arr=GIVEN_INPUT): 
    master_set = set(range(1, 10))
    output_arr = np.copy(input_arr)    
    num_zero_elem = np.count_nonzero(output_arr==0)
    num_iterations = 0
    need_to_drill_down = False
    possible_elems = [[set() for i in range(9)] for j in range(9)]
    impact_elems = np.zeros((9,9))
    num_total_updations = 0
    while num_zero_elem > 0:  
        num_updations = 0
        num_iterations += 1
        for i in range(9):
            for j in range(9):
                if output_arr[i, j] == 0:
                    rel_elem = get_row_col_sub_square_elem(i, j, output_arr) 
                    possible_elem = master_set - rel_elem
                    possible_elems[i][j] = possible_elem
                    if len(possible_elem) == 1:
                        output_arr[i, j] = possible_elem.pop()
                        num_updations += 1 
                else:
                    possible_elems[i][j] = set([output_arr[i, j]])
                
                impact_elems[i, j] = get_num_known_elem_vicinity_cur_elem(i, j, input_arr=output_arr)
                        
        num_zero_elem = np.count_nonzero(output_arr==0)

        num_total_updations += num_updations
        if ((num_zero_elem > 0) and (num_updations == 0)):
            break

    return (output_arr, num_zero_elem, num_total_updations, num_iterations, possible_elems, impact_elems)
 
```

```python
# In this scenario, we push several possible possibilities of the input array to a stack and these 
# are analysed separately, until we find one that satisfies the constraints of Sudoku.
def do_non_obvious_analysis(
    output_arr, impact_arr, possible_elems, run_mode='MIN_NUM_POSSIBILITIES_MIN_NUM_KNOWN_ELEM'):
    
    # DRILL DOWN
    print('\n## Not solved completely with obvious analysis. Digging further ##\n')
    num_stack_iterations = 0
    no_solution_found = True
      
    # Implementation of stack using lifoqueue
    stack = LifoQueue(maxsize=2000)

    sort_items_by_impact_and_push_to_stack(output_arr, impact_arr, possible_elems, stack, run_mode)
    while (not stack.empty()):

        num_stack_iterations += 1
        (row, col, elem, impact, output_arr) = stack.get().get_elements()
        copied_array = np.copy(output_arr)
        copied_array[row][col] = elem

        (output_arr, num_zero_elem, num_updations, num_iterations, possible_elems, impact_arr) = \
            do_obvious_analysis(copied_array)

        if num_zero_elem == 0:
            print('\n############# Solved with non obvious analysis ###################\n')
            no_solution_found = False
            break
        else:
            sort_items_by_impact_and_push_to_stack(output_arr, impact_arr, possible_elems, stack, run_mode)

    return (output_arr, num_stack_iterations, no_solution_found)
```

```python
def solve_sudoku(input_arr=GIVEN_INPUT, run_mode='MIN_NUM_POSSIBILITIES_MIN_NUM_KNOWN_ELEM', debug_output=False): 

    num_initial_unknowns = np.sum(input_arr == 0)
    no_solution_found = False
    num_stack_iterations = 0
    (output_arr, num_zero_elem, num_updations, num_iterations, possible_elems, impact_array) = \
        do_obvious_analysis(input_arr)
    
    if debug_output:
        print('\n################ Information regarding given input ##############\n') 
        print('Number of unknowns   : ', num_initial_unknowns)    

        print('\n####### Information regarding the initial obvious analysis ######\n')    
        print('Number of iterations : ', num_iterations)
        print('Number of updations  : ', num_updations)
        assert(num_initial_unknowns == (num_zero_elem + num_updations))
    if num_zero_elem ==0: 
        print('\n####### Solved completely with obvious analysis #################\n')
    elif not check_if_state_is_correct(output_arr):
        print('\n####### Wrong input,as no solution found #######\n')        
        no_solution_found = True
    else:
        
        (output_arr, num_stack_iterations, no_solution_found) = do_non_obvious_analysis(output_arr, 
                                                                                        impact_array, 
                                                                                        possible_elems,
                                                                                        run_mode)
            
    if debug_output:
        print("Number of stack iterations : ", num_stack_iterations)
        if not no_solution_found:
            print('\n####### Output ######\n')
            print('\n'.join([''.join(['{:4}'.format(item) for item in row]) 
      for row in output_arr]))
    return (output_arr, num_stack_iterations)
```
